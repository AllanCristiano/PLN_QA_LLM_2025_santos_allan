```python
import os
from langchain.document_loaders import PyPDFLoader, UnstructuredWordDocumentLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
from langchain.chains import RetrievalQA
from langchain.llms import HuggingFacePipeline
import transformers
import torch
```

```python
# Caminhos para os arquivos carregados
pdf_path = "doencas_respiratorias_cronicas.pdf"
docx_path = "DICIONARIO_DE_DADOS.docx"

# Carregar os documentos
loader_pdf = PyPDFLoader(pdf_path)
loader_docx = UnstructuredWordDocumentLoader(docx_path)

documents = loader_pdf.load() + loader_docx.load()
print(f"Total de páginas/documentos carregados: {len(documents)}")
```

```python
# Dividir os documentos em chunks
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=150)
docs = text_splitter.split_documents(documents)
print(f"Total de chunks de texto criados: {len(docs)}")

# Criar embeddings e o vector store (base de dados vetorial)
model_name = "sentence-transformers/all-MiniLM-L6-v2"
model_kwargs = {"device": "cuda" if torch.cuda.is_available() else "cpu"}
embeddings = HuggingFaceEmbeddings(model_name=model_name, model_kwargs=model_kwargs)

vectorstore = FAISS.from_documents(docs, embeddings)
print("Vector Store criado com sucesso.")
```

```python
# Função para carregar os pipelines dos modelos
def load_llm_pipeline(model_id):
    pipeline = transformers.pipeline(
        "question-answering",
        model=model_id,
        device=0 if torch.cuda.is_available() else -1, # Usa GPU se disponível
        framework="pt" # PyTorch
    )
    return pipeline

def load_generative_llm_pipeline(model_id):
    pipeline = transformers.pipeline(
        "text2text-generation",
        model=model_id,
        device=0 if torch.cuda.is_available() else -1,
        max_length=512,
        torch_dtype=torch.float32
    )
    return pipeline


# Carregando os modelos
print("Carregando modelo 1: distilbert-base-cased-distilled-squad...")
pipe_distilbert = load_llm_pipeline("distilbert-base-cased-distilled-squad")
llm_distilbert = HuggingFacePipeline(pipeline=pipe_distilbert)

print("Carregando modelo 2: deepset/roberta-base-squad2...")
pipe_roberta = load_llm_pipeline("deepset/roberta-base-squad2")
llm_roberta = HuggingFacePipeline(pipeline=pipe_roberta)

print("Carregando modelo 3: google/flan-t5-base...")
# Flan-T5 é um modelo generativo, então usamos uma pipeline diferente
pipe_flan_t5 = load_generative_llm_pipeline("google/flan-t5-base")
llm_flan_t5 = HuggingFacePipeline(pipeline=pipe_flan_t5)

print("Modelos carregados com sucesso.")
```

```python
print("\nIniciando o processo de Perguntas e Respostas...")

# Lista de perguntas definidas
questions = [
    # Perguntas do PDF
    "Quais são os principais fatores de risco para Doenças Respiratórias Crônicas (DRC)?",
    "Qual o tratamento farmacológico recomendado para rinite intermitente moderada a grave?",
    "Quais são os critérios para encaminhamento de um paciente com rinite alérgica para um especialista?",
    # Perguntas do DOCX
    "Qual a descrição da tabela LFCES004?",
    "Na tabela LFCES021 (TB_CARGA_HORARIA_SUS), o que significa o campo 'TP_SUS_NAO_SUS' e quais são seus domínios?",
    "Qual o nome da tabela no banco de produção federal que corresponde à tabela local LFCES037?"
]

# Dicionário para armazenar todos os resultados
results = {}

# Dicionários para organizar os modelos por tipo
extractive_pipelines = {"DistilBERT": pipe_distilbert, "RoBERTa": pipe_roberta}
qa_chain_flan_t5 = RetrievalQA.from_chain_type(
    llm=llm_flan_t5,
    chain_type="stuff",
    retriever=vectorstore.as_retriever()
)

# Criamos o retriever que buscará o contexto no vectorstore
retriever = vectorstore.as_retriever()

# Loop principal para processar cada pergunta com cada modelo
for i, question in enumerate(questions):
    print(f"\n--- Processando Pergunta {i+1}: '{question}' ---")
    results[question] = {}

    # 1. Primeiro, buscamos o contexto relevante (comum para todos os modelos)
    relevant_docs = retriever.get_relevant_documents(question)
    context = " ".join([doc.page_content for doc in relevant_docs])

    # 2. Executamos com os modelos EXTRATIVOS
    for model_name, pipeline in extractive_pipelines.items():
        print(f"Executando com o modelo: {model_name}")
        response_dict = pipeline(question=question, context=context)
        answer = response_dict.get('answer', 'Não foi possível encontrar uma resposta.')
        results[question][model_name] = answer
        print(f"Resposta ({model_name}): {answer}")

    # 3. Executamos com o modelo GENERATIVO
    print("Executando com o modelo: FLAN-T5")
    response = qa_chain_flan_t5.run(question)
    results[question][model_name] = response
    print(f"Resposta (FLAN-T5): {response}")

print("\n\n==============================================================================")
print("PROCESSO DE PERGUNTAS E RESPOSTAS CONCLUÍDO!")
print("==============================================================================")


# ==============================================================================
# EXIBIÇÃO FINAL DOS RESULTADOS
# ==============================================================================

print("\n\n--- RESUMO FINAL DAS RESPOSTAS ---\n")
for question, answers in results.items():
    print(f"PERGUNTA: {question}")
    for model_name, answer in answers.items():
        print(f"  - RESPOSTA ({model_name}): {answer}")
    print("-" * 50)
```
